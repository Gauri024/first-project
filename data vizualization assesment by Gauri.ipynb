```python
#step 1 import indian_foods.csv
df1= pd.read_csv('indian_foods.csv')
df1
```

```python
#replace the -1 with the mode of state column
state_mode=df1['state'].mode()[0]
state_mode
```

```python
df1 ['state']= df1['state'].replace(-1,state_mode)
df1
```

```python
#b)	Whatever the state you got in previous question, fill the corresponding region in the region column
df1.loc[df1['state']== state_mode,'region']=='West'
df1
```

```python
#c)	Replace the -1 with mean for cook time variable.
df1['cook_time']= df1['cook_time'].replace(-1,np.nan)
mean_time= np.round (df1['cook_time'].mean())
df1['cook_time']=df1['cook_time'].replace(np.nan, mean_time)
df1
```

```python
#d)	Most preferred flavours by number of customers using pie chart
flavours=df1['flavor_profile'].value_counts()
plt.pie(flavours, labels=flavours.index, autopct='%.f%%')
plt.title('Most Preferred Flavours by Customers')
plt.show()
```
